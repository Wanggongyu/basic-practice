画同心圆。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.arange(0, 2 * np.pi, 0.01) #角度集合
radius = np.arange(0, 5, 1) # 半径集合
```

```python
for r in radius:
    x = r * np.cos(theta) #对于每个半径，投影到x轴
    y = r * np.sin(theta) #投影到y轴
    plt.scatter(x, y, marker='.')
plt.axis('scaled') # 这样坐标轴不会被“压缩”
```
